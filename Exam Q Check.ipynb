## Can we emulate XOR with MLP?


```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-1,6,1000)
x2 = np.linspace(1,6,1000)

def H(z):
  if z >= 0:
    return 1
  else:
    return 0

def n1(x1,x2):
  z = x1+x2-1.5
  return H(z)

def n2(x1,x2):
  z = -x1-x2+0.5
  return H(z)

def n3(x1,x2):
  return H(x1+x2-1)



def perceptron(x1,x2):

  a4 = n1(x1,x2)
  a5 = n2(x1,x2)



  return H(a4+a5-1)

n = 1000
x = np.linspace(-1, 5, n)
y = np.linspace(-1, 5, n)

z=np.zeros((n,n))
for i in range(n):
  for j in range(n):
    z[i,j]=perceptron(x[i],y[j])

h = plt.contourf(x, y, z)
plt.axis('scaled')
plt.colorbar()
plt.show()


```
